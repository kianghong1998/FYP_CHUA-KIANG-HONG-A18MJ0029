```python
# Show accuracy of MLR model
from sklearn.metrics import r2_score, mean_squared_error
rmse = mean_squared_error(data['Measured Wind Speed'], data['Predicted Wind Speed (MLR)'], squared=False)
mse = mean_squared_error(data['Measured Wind Speed'], data['Predicted Wind Speed (MLR)'])
r_squared=r2_score(data['Measured Wind Speed'], data['Predicted Wind Speed (MLR)'])

print("Accuracy with raining")
print("The MLR model's r2 is ", r_squared*100, "%")
print("Mean squared error is ", mse)
print("Root mean squared error is ", rmse)
```

```python
# Show accuracy of ANN model
from sklearn.metrics import r2_score, mean_squared_error
rmse = mean_squared_error(data['Measured Wind Speed'], data['Predicted Wind Speed (ANN)'], squared=False)
mse = mean_squared_error(data['Measured Wind Speed'], data['Predicted Wind Speed (ANN)'])
r_squared=r2_score(data['Measured Wind Speed'], data['Predicted Wind Speed (ANN)'])

print("Accuracy with raining")
print("The ANN model's r2 is ", r_squared*100, "%")
print("Mean squared error is ", mse)
print("Root mean squared error is ", rmse)
```
